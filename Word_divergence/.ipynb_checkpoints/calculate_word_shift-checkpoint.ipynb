```python
texts_non = comment_cut[comment_cut['label']==0]['cut'].tolist()
texts_sex = comment_cut[comment_cut['label']==1]['cut'].tolist()
```

```python
import collections
def clean_text(doc_list):
    words = []
    for doc in doc_list:
        words += doc[1:-1].replace("'", "").split(", ")
    wordfreq_dict = collections.Counter(words)
    return wordfreq_dict
```

```python
freq_non = clean_text(texts_non)
freq_sex = clean_text(texts_sex)
```

```python
import shifterator as sh
```

```python
jsd_shift = sh.JSDivergenceShift(type2freq_1=freq_non,
                                 type2freq_2=freq_sex,
                                 weight_1=0.5,
                                 weight_2=0.5,
                                 base=2,
                                 alpha=1)
```
